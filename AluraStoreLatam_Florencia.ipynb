### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
ingreso_total1 = tienda["Precio"].sum()
print(ingreso_total1)

ingreso_total2 = tienda2["Precio"].sum()
print(ingreso_total2)

ingreso_total3 = tienda3["Precio"].sum()
print(ingreso_total3)

ingreso_total4 = tienda4["Precio"].sum()
print(ingreso_total4)
```

# 2. Ventas por categoría

En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

```python
tienda1_categorias = tienda.groupby('Categoría del Producto')['Precio'].sum().reset_index().sort_values('Precio',ascending=False)
print('Las ventas por Categoría de la Tienda 1:')
print(tienda1_categorias)

tienda2_categorias = tienda2.groupby('Categoría del Producto')['Precio'].sum().reset_index().sort_values('Precio',ascending=False)
print('Las ventas por Categoría de la Tienda 2:')
print(tienda2_categorias)

tienda3_categorias = tienda3.groupby('Categoría del Producto')['Precio'].sum().reset_index().sort_values('Precio',ascending=False)
print('Las ventas por Categoría de la Tienda 3:')
print(tienda3_categorias)

tienda4_categorias = tienda4.groupby('Categoría del Producto')['Precio'].sum().reset_index().sort_values('Precio',ascending=False)
print('Las ventas por Categoría de la Tienda 4:')
print(tienda4_categorias)
```

# 3. Calificación promedio de la tienda


En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.


```python
calificacion_general_1 = tienda['Calificación'].mean()
print(f'La calificacion promedio de los clientes para la tienda 1 es: {calificacion_general_1}')

calificacion_general_2 = tienda2['Calificación'].mean()
print(f'La calificacion promedio de los clientes para la tienda 2 es: {calificacion_general_2}')

calificacion_general_3 = tienda3['Calificación'].mean()
print(f'La calificacion promedio de los clientes para la tienda 3 es: {calificacion_general_3}')

calificacion_general_4 = tienda4['Calificación'].mean()
print(f'La calificacion promedio de los clientes para la tienda 4 es: {calificacion_general_4}')
```

```python
#Para conocer la satisfacción del cliente con los productos vendidos:
calif_categoriat1 = tienda.groupby('Categoría del Producto')['Calificación'].mean().sort_values(ascending=False)
print('La calificacion de los clientes por categoria de la tienda 1:')
print(calif_categoriat1)

calif_categoriat2 = tienda2.groupby('Categoría del Producto')['Calificación'].mean().sort_values(ascending=False)
print('La calificacion de los clientes por categoria de la tienda 2:')
print(calif_categoriat2)

calif_categoriat3 = tienda3.groupby('Categoría del Producto')['Calificación'].mean().sort_values(ascending=False)
print('La calificacion de los clientes por categoria de la tienda 3:')
print(calif_categoriat3)

calif_categoriat4 = tienda4.groupby('Categoría del Producto')['Calificación'].mean().sort_values(ascending=False)
print('La calificacion de los clientes por categoria de la tienda 4:')
print(calif_categoriat4)
```

# 4. Productos más y menos vendidos

```python
#En este paso, debes identificar qué productos fueron
#los más vendidos y los menos vendidos en cada tienda.
#Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

masvendidos_tienda1 = tienda.groupby('Producto')[['Precio']].sum().sort_values('Precio',ascending=False).reset_index()
print('Los productos mas vendidos de la tienda 1 fueron:')
print(masvendidos_tienda1.head())

masvendidos_tienda2 = tienda2.groupby('Producto')[['Precio']].sum().sort_values('Precio',ascending=False).reset_index()
print('\nLos productos mas vendidos de la tienda 2 fueron:')
print(masvendidos_tienda2.head())

masvendidos_tienda3 = tienda3.groupby('Producto')[['Precio']].sum().sort_values('Precio',ascending=False).reset_index()
print('\nLos productos mas vendidos de la tienda 3 fueron:')
print(masvendidos_tienda3.head())

masvendidos_tienda4 = tienda4.groupby('Producto')[['Precio']].sum().sort_values('Precio',ascending=False).reset_index()
print('\nLos productos mas vendidos de la tienda 4 fueron:')
print(masvendidos_tienda4.head())
```

```python
#los menos vendidos en cada tienda:

producto_tienda1 = tienda.groupby('Producto')[['Precio']].sum().sort_values('Precio').reset_index()
print('Los productos menos vendidos de la tienda 1 fueron:')
print(producto_tienda1.head())


producto_tienda2 = tienda2.groupby('Producto')[['Precio']].sum().sort_values('Precio').reset_index()
print('\nLos productos menos vendidos de la tienda 2 fueron:')
print(producto_tienda2.head())

producto_tienda3 = tienda3.groupby('Producto')[['Precio']].sum().sort_values('Precio').reset_index()
print('\nLos productos menos vendidos de la tienda 3 fueron:')
print(producto_tienda3.head())

producto_tienda4 = tienda4.groupby('Producto')[['Precio']].sum().sort_values('Precio').reset_index()
print('\nLos productos menos vendidos de la tienda 4 fueron:')
print(producto_tienda4.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

producto_tienda1 = tienda.groupby('Producto')[['Precio']].sum().sort_values('Precio').reset_index()

plt.figure(figsize=(10, 10))
# Ahora 'Producto' y 'Precio' son columnas claras para el gráfico
sns.barplot(y='Producto', x='Precio', data=producto_tienda1, color='#3C2C73')

plt.title('Comparación de venta de cada Producto - Tienda 1')
plt.xlabel('Ventas Totales ($)')
plt.ylabel('Productos')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.ticklabel_format(style='plain', axis='x')
plt.tight_layout()
plt.show()
```

# 5. Envío promedio por tienda

```python
#En este paso, debes calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.  Sabiendo que el cliente paga el envío.

envio_promedio_tienda1 = tienda['Costo de envío'].mean()
print(f'El costo de envio promedio de la tienda 1 es: {envio_promedio_tienda1}')

envio_promedio_tienda2 = tienda2['Costo de envío'].mean()
print(f'El costo de envio promedio de la tienda 2 es: {envio_promedio_tienda2}')

envio_promedio_tienda3 = tienda3['Costo de envío'].mean()
print(f'El costo de envio promedio de la tienda 3 es: {envio_promedio_tienda3}')

envio_promedio_tienda4 = tienda4['Costo de envío'].mean()
print(f'El costo de envio promedio de la tienda 4 es: {envio_promedio_tienda4}')
```

$363M, lo que demuestra una pérdida de competitividad en productos de alto valor.
